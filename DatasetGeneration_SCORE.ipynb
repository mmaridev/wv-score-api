```python
import pandas as pd
```

```python
import random
```

```python
data = {
    "age": [50], "sex": ["M"], "cholesterol": [6], "max_press": [120], "smokes": [True]
}
```

```python
for i in range(999):    
    data["age"].append(random.randint(40,100))
    data["sex"].append(random.choice(["M", "F"]))
    data["cholesterol"].append(random.randint(2, 10))
    data["max_press"].append(random.randint(80, 180))
    data["smokes"].append(bool(random.getrandbits(1)))
```

```python
data
```

```python
df = pd.DataFrame.from_dict(data)
df.head()
```

```python
df.tail()
```

```python
df.to_csv("dataset.csv")
```
